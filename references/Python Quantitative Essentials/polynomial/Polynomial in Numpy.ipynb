```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

###  Method 1:  Polynomials as arrays (Like MATLAB)

Polynomials are represented as arrays in order of descending coefficients

For example, the array [5, -3, 2, 1] represents
$$5x^3 - 3x^2 + 2x + 1$$

There are suites of function for manipulating these polynomials.
- polyval --  Numerically evaluate at point(s)
- roots   --  Find toor
- polyder --  Derivative
- poltiny --  Integral

```python
#  Create an array representing a polynomial
P1 = [5, -3, 2, 1]

#  Calculate and print the roots
R1 = np.roots(P1)
print(R1)

#  Use polyval to evaulate an a numer of x-values
x = np.linspace(-1, 1, 100)
y = np.polyval(P1, x)

#  Plot the function
plt.plot(x, y, 'k')
plt.plot(R1[-1], 0, 'o')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y)')
```

### Poly1d

Coefficient order is the same as in above

Poly1d returns an object

Some class atributes are calculated on creation
- roots

Member functions are
- deriv
- integ

```python
x = np.linspace(-1, 1, 100)

#  Create a poly1d object
P2 = np.poly1d(  [5, -3, 2, 1] )

#   Plot it
plt.plot(x, P2(x), 'k')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y)')

#  Print the roots
P2.roots

#  Print out the coefficients
P2.deriv().coef
```

###  Polynomial class

<a href="https://numpy.org/doc/stable/reference/routines.polynomials.html">Documentation</A>

Coefficients are in ascending order

There are numerous function for arithmetic (add, muiltply, etc.) calculus, and building polynomials in different ways.

```python
#  [5, -3, 2, 1]  Remember for this, the order of the coefficients is reversed.
#  Create a Polynomial object
P3 = np.polynomial.Polynomial( [1, 2, -3, 5] )

x = np.linspace(-1, 1, 100)

#  Plot it
plt.plot(x, P3(x), 'k')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y)')

#  print the roots
print( P3.roots() )
```

###  An example multiplying polynomials
If we wish to multiply two polynomials together, for example, 
$$2x^2 + x$$
and
$$-x^3 + 2x^2 + 1$$

```python
#  Define the two polynomals
A = np.polynomial.Polynomial( [0, 1, 2] )
B = np.polynomial.Polynomial( [1, 0, 2, -1] )

#  Find the product
C = np.polynomial.polynomial.polymul(A, B)
print(C)
```
