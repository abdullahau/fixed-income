```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf

%matplotlib inline
```

```python
#  Import S^P 500 data from CSV file.
#  Format:  Date, Open, High, Low, Close, Adjusted Close, Volume
SPX = pd.read_csv('SPX.csv')

#  For the demo, we doon't want the whole set.  We'll grab the last 30 points only.
SPX = SPX.tail(30)
```

```python
#  Plot it with Pandas
#SPX.plot(x = 'Date', y = 'Close')

#  Convert the Dates from ASCII text to datetime objects
SPX.iloc[:,0] = pd.to_datetime( SPX.iloc[:, 0], format = '%Y-%m-%d')

#  Set the dates as the index
SPX = SPX.set_index( pd.DatetimeIndex(SPX['Date']))

#  Generate the candlestick plot, along with 5 and 10 day moving averages
mpf.plot(SPX, type = 'candle', mav=(5, 10))
```

```python
#  Let's calculate the moving averages with Pandas using the rolling function
SPX['MA5'] = SPX['Close'].rolling(5).mean()
SPX['MA1-'] = SPX['Close'].rolling(10).mean()
print(SPX.head(10))
```
