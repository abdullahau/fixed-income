### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Function

```python
def f(x, a, b, c):
  return a * x**2 + b*x + c

xlist = np.linspace(-10, 10, num=1000)
# xlist = np.arange(-10, 10.1, 0.1)


ylist = f(xlist, 3, 1, 4)
```

### Plot

```python
plt.figure(num=0, dpi=120)
plt.plot(xlist, ylist, label="$f(x)$")
plt.plot(xlist, ylist**(1/2), "--g", label="$f(x)^{1/2}$")
plt.title('Plotting Example - Quadratic Function')
plt.xlabel('Distance / ft')
plt.ylabel('Height / ft')
plt.legend()
```
