### S&P 500

```python
sp500 = yf.Ticker('^GSPC')
hist = sp500.history(period='50y', interval='1d')

plt.figure(num=0, dpi=120)
plt.plot(hist['Close'])
plt.title('S&P 500 Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
```
